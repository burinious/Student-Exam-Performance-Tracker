```python
def compute_average(scores):
    count = 0
    total = 0

    for score in scores:
        if score is not None:
            total += score
            count += 1

    if count > 0:
        return total / count
    else:
        return 0

def compute_grade(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

def compute_final_grade(grades):
    if all(grade == "A" or grade == "B" or grade == "C" for grade in grades):
        return "Pass"
    elif any(grade == "D" for grade in grades):
        return "Retake Exam"
    elif any(grade == "F" for grade in grades):
        return "Fail"
    else:
        return ""

def compute_overall_average(students):
    total_score = 0
    count = 0

    for student in students:
        chem_avg = compute_average(student["courses"]["Chem"])
        bio_avg = compute_average(student["courses"]["Bio"])
        phy_avg = compute_average(student["courses"]["Phy"])
        math_avg = compute_average(student["courses"]["Math"])

        total_score += (chem_avg + bio_avg + phy_avg + math_avg) / 4
        count += 1

    if count > 0:
        return total_score / count
    else:
        return 0

    
def generate_performance_record(student):
    chem_scores = student["courses"]["Chem"]
    bio_scores = student["courses"]["Bio"]
    phy_scores = student["courses"]["Phy"]
    math_scores = student["courses"]["Math"]

    chem_average = compute_average(chem_scores)
    bio_average = compute_average(bio_scores)
    phy_average = compute_average(phy_scores)
    math_average = compute_average(math_scores)

    chem_grade = compute_grade(chem_average)
    bio_grade = compute_grade(bio_average)
    phy_grade = compute_grade(phy_average)
    math_grade = compute_grade(math_average)

    overall_avg = (chem_average + bio_average + phy_average + math_average) / 4
    overall_grade = compute_final_grade([chem_grade, bio_grade, phy_grade, math_grade])
    
   

    return f"{student['name']}          | {chem_grade}         | {bio_grade}         | {phy_grade}         | {math_grade}         | {overall_avg}         | {overall_grade}"

    if overall_avg >= 80:
        return "A"
    elif overall_avg >= 70:
        return "B"
    elif overall_avg >= 60:
        return "C"
    elif overall_avg >= 50:
        return "D"
    else:
        return "F"

def generate_exam_results_table(students):
    header = "Student name | Chem Grade | Bio Grade | Phy Grade | Math Grade | Overall Avg | Exam Status\n"
    table = header + "-" * len(header) + "\n"

    for student in students:
        performance_record = generate_performance_record(student)
        table += performance_record + "\n"

    return table

def main():
    students = []

    while True:
        name = input("Enter student name (or 'q' to quit): ")
        if name.lower() == "q":
            break

        courses = {}

        while True:
            course = input("Enter course name (or 'q' to finish adding courses): ")
            if course.lower() == "q":
                break

            scores = input("Enter exam scores separated by space (or 'q' to finish adding scores): ").split()

            try:
                scores = [float(score) if score != "-" else None for score in scores]
                courses[course] = scores
            except ValueError:
                print("Invalid score. Please enter valid numbers.")

        student = {
            "name": name,
            "courses": courses
        }

        students.append(student)

    print("\nPerformance Records:")
    print(generate_exam_results_table(students))

if __name__ == "__main__":
    main()
```

## BONUS: Generate a table of exam results for each subject

```python
def compute_average(scores):
    count = 0
    total = 0

    for score in scores:
        if score is not None:
            total += score
            count += 1

    if count > 0:
        return total / count
    else:
        return 0

def compute_grade(score):
    if 80 <= score <= 100:
        return "A"
    elif 70 <= score < 80:
        return "B"
    elif 60 <= score < 70:
        return "C"
    elif 50 <= score < 60:
        return "D"
    else:
        return "F"

def compute_final_grade(grades):
    if all(grade == "A" or grade == "B" or grade == "C" for grade in grades):
        return "Pass"
    elif any(grade == "D" for grade in grades):
        return "Retake Exam"
    elif any(grade == "F" for grade in grades):
        return "Fail"
    else:
        return ""

def generate_subject_results_table(students, subject):
    header = f"Student name | {subject} Score | {subject} Grade\n"
    table = header + "-" * len(header) + "\n"

    for student in students:
        subject_scores = student["courses"][subject]
        subject_average = compute_average(subject_scores)
        subject_grade = compute_grade(subject_average)
        table += f"{student['name']}          | {subject_average:.2f}         | {subject_grade}\n"

    return table

def generate_exam_results_table(students):
    chem_table = generate_subject_results_table(students, "Chem")
    bio_table = generate_subject_results_table(students, "Bio")
    phy_table = generate_subject_results_table(students, "Phy")
    math_table = generate_subject_results_table(students, "Math")

    return (
        f"Chemistry Results\n{chem_table}\n\n"
        f"Biology Results\n{bio_table}\n\n"
        f"Physics Results\n{phy_table}\n\n"
        f"Math Results\n{math_table}"
    )

def main():
    students = []

    while True:
        name = input("Enter student name (or 'q' to quit): ")
        if name.lower() == "q":
            break

        courses = {}

        while True:
            course = input("Enter course name (or 'q' to finish adding courses): ")
            if course.lower() == "q":
                break

            scores = input("Enter exam scores separated by space (or 'q' to finish adding scores): ").split()

            try:
                scores = [float(score) if score != "-" else None for score in scores]
                courses[course] = scores
            except ValueError:
                print("Invalid score. Please enter valid numbers.")

        student = {
            "name": name,
            "courses": courses
        }

        students.append(student)

    print("\nPerformance Records:")
    print(generate_exam_results_table(students))

if __name__ == "__main__":
    main()

    
```
